决策树算法分为ID3、C4.5、CART等算法
$$H(A,F^{(j)}=f)=\frac{|A_1|}{|A|}H(A_1)+\frac{|A_2|}{|A|}H(A_2)$$
信息增益(**ID3算法**采用了)：
$$Gain(A,F^{(j)}=f)=H(A)-H(A,F^{(j)}=f)$$
增益率(Gain Ratio,**C4.5算法**采用了)
$$GainRatio(A,F^{(j)})=\frac{Gain(A,F^{(j)})}{SplitInfo(F^{(j)})}$$
$$SplitInfo(F^{(j)})=-\sum\frac{|A_i|}{A}\log_2\frac{|A_i|}{|A|}$$
基尼指数(**CART算法**采用了)
$$Gini(p)=\sum_{k=1}^{K}p_k(1-p_k)=1-\sum_{k=1}^{K}p_k^2$$

随机森林（Random Forests，RF）算法：构建多颗决策树，从样本集中**有放回**地重复随机抽样生成新的样本集合，然后**无放回**地随机选择若干特征生成一颗决策树，最终每棵树的结果投票决定最终分类

需要预先确定两个参数，构建决策树的个数n和构建时参考的特征数k，k一般取$\log_2K$

n_estimators,max_features

平均准确率$$average\_accuracy=\frac{\left ( \frac{TP}{TP+FN} + \frac{TN}{FP+TN} \right)}{2}$$

$$F_\beta=(1+\beta^2)\frac{precision\times recall}{(\beta^2 \times precision)+recall},\beta一般取2或0.5$$
