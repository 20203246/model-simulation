$$f(x)=W\cdot x=\sum_{i=0}^n\omega^{(i)}\cdot x^{(i)}, x^0=1$$
$$s_i^2(W)=(y_i-f(x_i))^2=(y_i-W\cdot x_i)^2$$
最小化：
$$L(W)=s_1^2(W)+s_2^2(W)+\dots+s_m^2(W)=(Y-W\overline{X})(Y-W\overline{X})^T$$
求导...

$$\hat{W}^T=(\overline{X}\overline{X}^T)^{-1}\overline{X}Y^T$$

```python
temperatures = [10,15,20,25,30,35]
insects = [39,42,50,57,58,65]

import numpy as np
def lease_square(X, Y):
    '''
    :param X:矩阵，样本特征函数
    :param Y:矩阵，标签向量
    :return :矩阵，回归系数
    '''
    return (X * X.T).I * X * Y.T
X = np.mat([[1,1,1,1,1,1],temperatures])
Y = np.mat(insects)
W = lease_square(X,Y)
print(W)
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.scatter(temperatures,insects,c='g',label='原始点')
x1 = np.linspace(10,40,100)
y1 = W[1,0] * x1 + W[0,0]
plt.plot(x1,y1,c='r',label='拟合直线')
plt.legend()
plt.show()
```
